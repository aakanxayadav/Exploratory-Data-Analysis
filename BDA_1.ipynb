```python
#  Sales vs. Expenses over Time

import matplotlib.pyplot as plt
df.plot(kind='line',x='Month',y=['Sales ','Expenses '], figsize=(10,6))
plt.xlabel('Month')
plt.ylabel('Amount')
_ = plt.title('Sales vs. Expenses over Time')
```

```python
# Net Profit vs EPS in Rs

import matplotlib.pyplot as plt
df.plot(kind='scatter', x='Net Profit ', y='EPS in Rs', c='Tax %', cmap='viridis')
plt.xlabel('Net Profit')
_ = plt.ylabel('EPS in Rs')
```
